### Built function :-

```python
name = "yasmeen mohamed mohamed roshdy"

#Upper case
print(name.upper())

#lower case
print(name.lower())

#replace
print(name.replace("yasmeen" , "soo"))

#capitalize  (first letter capital)
print(name.capitalize())  

#title (first letter of each word capital)
print(name.title())

```
