### built functions :-

```python
name = "depi learning depi depi"
#........................count.......................

print(name.count("depi"))

#........................split.........................

a= name.split()   #بتقسم الجملة لكلمات
print(a)

 #.......................join.........................

b=" ".join(a)  #جمع الجملة وافصل بين كل كلمة بمسافة
print(b)

c = "".join(a) #جمع الجملة بدون مسافات
print(c)

#......................startswith....................

d = name.startswith("depi")  #بتشيك اذا الجملة بتبدأ بالكلمة المحددة
print(d)

#......................endswith......................

e = name.endswith("depi")    #بتشيك اذا الجملة بتنتهي بالكلمة المحددة
print(e)
```

##### append(end):-

```python
list2=[1,2,3,4]
list2.append(7)
print(list2)
```

##### insert :-

```python
list3=[1,2,3,4]
list3.insert(0,10)  #بتضيف عنصر في مكان معي
print(list3)
```

##### extend :-

```python
list2.extend(list3) #بضم ليست لتانية
print(list2)
print(list3)
```

##### del[position] :-

```python
del list3[0]  #بحذف عنصر من الليست حسب المكان
print(list3)
```

##### remove (value):-

```python
l = ["a", "b", "b","b","c", "d", "e"]
l.remove("c")  #بحذف عنصر من الليست حسب القيمة
print(l)
l.remove("b") #بحذف اول عنصر مطابق للقيمة
print(l)
```

##### pop :-

```python
l2 = ["a", "b", "b","b","c", "d", "e"]
a = l2.pop()  #بحذف اخر عنصر في الليست
print(l2)
print(a) #بترجعلي العنصر اللي اتمسح اذا خزنته في فاريبل
```
